```python
import requests
import pandas as pd

headers = {
	"X-RapidAPI-Key": "55c5f05713mshe9a9f9981401e48p19f947jsncb188421bf55",
	"X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
}

def get_codes():
    
    location = input('Insert location to find airport(s) information: ')
    print()
    
    url = "https://aerodatabox.p.rapidapi.com/airports/search/term"
    querystring = {"q":location.lower(), "limit":"250"}

    response = requests.get(url, headers=headers, params=querystring)
    codes = response.json()['items'] if 'items' in response.json() else location

    return codes


def get_flights():
    start = input('Enter a first date & time in formart (YYYY-MM-DDTHH:mm): ')
    end = input('Enter a second date & time in formart (YYYY-MM-DDTHH:mm): ')
    icao = input('Enter the icao (or one of the icao(s)) you have from your previous search: ')
    print()
    
    url = f"https://aerodatabox.p.rapidapi.com/flights/airports/icao/{icao}/{start}/{end}"
    querystring = {"withLeg":"true","withCancelled":"true","withCodeshared":"true","withCargo":"true","withPrivate":"true","withLocation":"false"}
    
    response = requests.request("GET", url, headers=headers, params=querystring)

    response_data = response.json()
    departures = response_data['departures']
    arrivals = response_data['arrivals']

    df_departures = pd.json_normalize(departures)
    df_arrivals = pd.json_normalize(arrivals)

    df = pd.concat([df_departures, df_arrivals])
    return df
```
